```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
TRAIN_ANNOT_DIR = './dataset/mask/train'
TEST_ANNOT_DIR = './dataset/mask/test'
```

```python
def split_images(impath='./dataset/mask/train'):
    imlist = os.listdir(impath)
    output_dir = os.path.join("./annot_split/" , impath.split('/')[-1])
    

    for imgname in imlist:
        image = cv2.imread(os.path.join(impath,imgname)) 
        unique_values = np.unique(image)
        for value in unique_values:
            os.makedirs(f"{output_dir}/{value}/", exist_ok=True)
            mask = (image == value)
            
            new_image = np.zeros_like(image)
            new_image[mask] = value

            # Save the new image
            output_path = f"{output_dir}/{value}/{imgname.split('.')[0]}_{value}.png"
            cv2.imwrite(output_path, new_image)
```

```python
split_images(TRAIN_ANNOT_DIR)
split_images(TEST_ANNOT_DIR)
```

```python
def count_images(path):
    return {val:len(os.listdir(os.path.join(path,str(val)))) for val in range(12)}
```

```python
traincount = count_images('annot_split/train/')
testcount = count_images('annot_split/test/')
```

```python
def plot_samples(root_dir):
    # Number of rows and columns in the plot
    num_rows = 12
    num_cols = 5

    # Create a figure and axis for plotting
    _, axes = plt.subplots(num_rows, num_cols, figsize=(10, 20))

    # Loop through each class directory
    for class_idx in range(num_rows):
        class_dir = os.path.join(root_dir, str(class_idx))
        image_files = os.listdir(class_dir)
        image_samples = random.sample(image_files, num_cols)
        # Loop through each sample image in the class
        for col_idx, image_file in enumerate(image_samples):
            image_path = os.path.join(class_dir, image_file)
            image = np.array(plt.imread(image_path) * 255)
            print(np.unique(image))
            
            # Plot the image
            axes[class_idx, col_idx].set_title(image_file, fontsize=8)
            axes[class_idx, col_idx].imshow(image)
            axes[class_idx, col_idx].axis('off')

    # Adjust layout and display the plot
    # plt.tight_layout()
    # plt.show()
```

```python
plot_samples('annot_split/train/')
```
